```python
#importing packages
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
#input data
dataframe = pd.read_csv("C:/Users/DELL/Downloads/Input.xlsx - Sheet1.csv")
```

```python
url=dataframe["URL"]
url_id=dataframe["URL_ID"]
```

```python
all_title=[]
all_body=[]

for i in url:
    print (i)
    response = requests.get(i)
    soup = bs(response.content, 'html.parser')
    title=soup.find("head").find("title")

    body_text= soup.find("div",attrs={"class","td-post-content tagdiv-type"})
    if title is not None:
        all_title.append(title.text)
    else:
        all_title.append('Not Found')
    if body_text is not None:
        all_body.append(body_text.text)
    else:
        if len(soup.find_all("div",attrs={"class","tdb-block-inner td-fix-index"}))>15:
            body_text= soup.find_all("div",attrs={"class","tdb-block-inner td-fix-index"})[14]
            if body_text is not None:
                all_body.append(body_text.text)
        else:
            all_body.append("Not Found")


```

```python
# creating dataframe
df_1=pd.DataFrame({'URL':dataframe["URL"],'Title':all_title,'Body':all_body})
df_1.head()
```

```python
len(df_1)
```

```python
df_1.drop([24,37], inplace = True)
len(df_1)
```

```python
# total word list

words_list=[]
for i in df_1['Body']:
    first=i

    tokens = word_tokenize(first)
    # convert to lower case
    tokens = [w.lower() for w in tokens]
    # remove punctuation from each word

    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in tokens]
    # remove remaining tokens that are not alphabetic
    words = [word for word in stripped if word.isalpha()]
    # filter out stop words

    stop_words = set(stopwords.words('english'))
    words = [w for w in words if not w in stop_words]
    words_list.append(words)
```

```python
total_word_count=[]
for i in words_list:
    count=0
    for j in i:
        count=count+len(j)
    total_word_count.append(count)
```

```python
stopword_list=os.listdir("C:/Users/DELL/Desktop/Python/BlackCoffer/StopWords")
```

```python
#stop word list

for i in stopword_list:
    loc="C:/Users/DELL/Desktop/Python/BlackCoffer/StopWords"+"/"+ i
    with open(loc) as f:
        contents1 = f.readlines()
    stop_string1=' '.join(contents1)
    from nltk.tokenize import word_tokenize
    tokens = word_tokenize(stop_string1)
    # convert to lower case
    tokens = [w.lower() for w in tokens]
    # remove all tokens that are not alphabetic
    s_words = [word for word in tokens if word.isalpha()]
len(s_words)
```

```python
# positive word list

with open("C:/Users/DELL/Desktop/Python/BlackCoffer/MasterDictionary/positive-words.txt") as p:
    positive1 = p.readlines()
    posi_string1=' '.join(positive1)
from nltk.tokenize import word_tokenize
tokens = word_tokenize(posi_string1)
# remove all tokens that are not alphabetic
positive_words = [word for word in tokens if word.isalpha()]
len(positive_words)
```

```python
# positive word list after comparing with stopwords

positive_words1 = [i for i in positive_words if i not in s_words]
len(positive_words1)
```

```python
#positive score

positive_score_list=[]
for word in words_list:
    positive_score = len([i for i in word if i in positive_words1])
    positive_score_list.append(positive_score)
```

```python
# negetive word list

with open("C:/Users/DELL/Desktop/Python/BlackCoffer/MasterDictionary/negative-words.txt") as p:
    negetive1 = p.readlines()
    nege_string1=' '.join(negetive1)
from nltk.tokenize import word_tokenize
tokens = word_tokenize(nege_string1)
# remove all tokens that are not alphabetic
negetive_words = [word for word in tokens if word.isalpha()]
len(negetive_words)
```

```python
# negetive word list after comparing with stopwords

negetive_words1 = [i for i in negetive_words if i not in s_words]
len(negetive_words1)
```

```python
# negetive score

negetive_score_list=[]
for word in words_list:
    negetive_score = len([i for i in word if i in negetive_words1])
    negetive_score_list.append(negetive_score)
```

```python
#Polarity Score

polarity_score=[]
for i in range(len(words_list)):
    Polarity_Score = (positive_score_list[i]-negetive_score_list[i])/(positive_score_list[i]+negetive_score_list[i] + 0.000001)
    polarity_score.append(Polarity_Score)
max(polarity_score)
```

```python
min(polarity_score)
```

```python
# Subjectivity Score
subjectivity_score=[]
for i in range(len(words_list)):
    Subjectivity_Score = (positive_score_list[i] +negetive_score_list[i])/(len(words_list[i])+0.000001)
    subjectivity_score.append(Subjectivity_Score)
min(subjectivity_score)
```

```python
max(subjectivity_score)
```

```python
 # split into sentences
sentences_list=[]
for i in df_1['Body']:
    first=i
    from nltk import sent_tokenize
    sentences = sent_tokenize(first)
    sentences_list.append(sentences)
len(sentences_list)
```

```python
# count of uncleaned words
word_list1=[]
for i in df_1['Body']:
    first=i
    from nltk.tokenize import word_tokenize
    tokens = word_tokenize(first)
    word_list1.append(tokens)
len(word_list1[0])
```

```python
# Average Sentence Length
average_sentence_length = []
for i in range(len(words_list)):
    Average_Sentence_Length=len(words_list[i]) /len(sentences_list[i])
    average_sentence_length.append(Average_Sentence_Length)
```

```python
average_sentence_length[0]
```

```python
#Average Number of Words Per Sentence (using unclean words)

average_no_words = []
for i in range(len(words_list)):
    Average_no_Length=len(word_list1[i]) /len(sentences_list[i])
    average_no_words.append(Average_no_Length)
```

```python
average_no_words[0]
```

```python
# Syallable Word Count

from nltk.tokenize import SyllableTokenizer

syllable_list=[]
for i in words_list:
    syllable_w_list=[]
    for j in i:
        SSP = SyllableTokenizer()
        a=SSP.tokenize(j)
        syllable_w_list.append(a)
    syllable_list.append(syllable_w_list)

```

```python
syllable_list[0][0]
```

```python
syllable_count=[]
for i in syllable_list:
    count=0
    for j in i:
        count=count+len(j)
    syllable_count.append(count)
```

```python
len(syllable_count)
```

```python
# Complex Word Count

complex_word_count=[]
for i in range(len(syllable_list)):
    complex_count=0
    for j in (syllable_list[i]):
        if len(j)>2:
            complex_count=complex_count+1
    complex_word_count.append(complex_count)
```

```python
len(complex_word_count)
```

```python
#Percentage of Complex words
Percentage_complex_words=[]
for i in range(len(words_list)):
    Percentage_complex_words.append(complex_word_count[i] /len(words_list[i]))
```

```python
Percentage_complex_words[0]
```

```python
# Fog Index

Fog_Index=[]
for i in range(len(words_list)):
    Fog_Index1 = 0.4 * (average_sentence_length[i] + Percentage_complex_words[i])
    Fog_Index.append(Fog_Index1)
```

```python
Fog_Index[0]
```

```python

#Sum of the total number of characters in each word/Total number of words
total_char=[]
for i in range(len(words_list)):
    count=0
    for j in words_list[i]:
        count=count+len(j)
    total_char.append(count)
```

```python
# Average Word Length
Average_word_length=[]
for i in range(len(words_list)):
    a=total_char[i]/len(words_list[i])
    Average_word_length.append(a)
```

```python
#Personal Pronouns

import re
personal_pronouns=[]
pronounRegex = re.compile(r'\b(I|we|my|ours|(?-i:us))\b',re.I)
for i in range(len(sentences_list)):
    count=0
    for j in sentences_list[i]:
        pronouns = pronounRegex.findall(j)
        count=len(pronouns)+count
    personal_pronouns.append(count)
```
